```python
"""
Code to change current working directory to the our code's and data's directory.
"""
from google.colab import drive
import os
import cv2
from sklearn.preprocessing import StandardScaler
import numpy as np
import pickle
import pandas as pd

drive.mount("/content/drive")
os.chdir("drive/MyDrive/Colab Notebooks/ML Project")
print("Current Working Directory:", os.getcwd())
train_dir = 'chest_xray/train'
test_dir = 'chest_xray/test'

img_dim = 32 # IMPORTANT
```

```python
def build_dataframe(path):
  cnt = 0
  dataArray = []
  for item in os.listdir(path):
    cnt += 1
    print(cnt)
    img = cv2.imread(path+"/"+item)
    img = cv2.resize(img, (img_dim, img_dim), interpolation = cv2.INTER_AREA).reshape((-1))
    dataArray.append(img)
  return pd.DataFrame(np.array(dataArray))

def build_dataframe_with_flip(path):
  dataArray = []
  cnt = 0
  for item in os.listdir(path):
    cnt += 1
    print(cnt)
    img = cv2.imread(path+"/"+item)
    img_o = cv2.resize(img, (img_dim, img_dim), interpolation = cv2.INTER_AREA).reshape((-1))
    dataArray.append(img_o)
    img_f = cv2.flip(img, 1)
    img_f = cv2.resize(img_f, (img_dim, img_dim), interpolation = cv2.INTER_AREA).reshape((-1))
    dataArray.append(img_f)
  return pd.DataFrame(np.array(dataArray))
```

**Imbalanced Normal Train Set**

```python
normal_train_df = build_dataframe(train_dir + '/NORMAL')
f = open("Imbalance_normal_train_rgb_df_" + str(img_dim) + ".pkl", "wb")
pickle.dump(normal_train_df, f)
f.close()
```

**Balanced Normal Train Set**

```python
b_normal_train_df = build_dataframe_with_flip(train_dir + '/NORMAL')
f = open("Balance_normal_train_rgb_df_" + str(img_dim) + ".pkl", "wb")
pickle.dump(b_normal_train_df, f)
f.close()
```

**Pneumonia Train Set**

```python
pneumonia_train_df = build_dataframe(train_dir + '/PNEUMONIA')
f = open("Pneumonia_train_rgb_df_" + str(img_dim) + ".pkl", "wb")
pickle.dump(pneumonia_train_df, f)
f.close()
```

**Test Set**

```python
normal_test_df = build_dataframe(test_dir + '/NORMAL')
pneumonia_test_df = build_dataframe(test_dir + '/PNEUMONIA')
test_df = pd.concat([normal_test_df, pneumonia_test_df])
f = open("Test_rgb_df_" + str(img_dim) + ".pkl", "wb")
pickle.dump(test_df, f)
f.close()
```
