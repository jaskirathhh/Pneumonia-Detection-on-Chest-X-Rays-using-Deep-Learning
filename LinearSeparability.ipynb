```python
import pandas as pd
import pickle
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

normal_df = None
with open('/content/drive/MyDrive/Colab Notebooks/ML Project/Imbalance_normal_train_rgb_df_32.pkl', 'rb') as f:
  normal_df = pickle.load(f)

pneumonia_df = None
with open('/content/drive/MyDrive/Colab Notebooks/ML Project/Pneumonia_train_rgb_df_32.pkl', 'rb') as f:
  pneumonia_df = pickle.load(f)
```

```python
scaler = StandardScaler()
df = pd.concat([normal_df, pneumonia_df])
df = pd.DataFrame(scaler.fit_transform(df))
df['Label'] = pd.Series(np.concatenate((np.zeros(1350), np.ones(3885))))
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
pca = PCA(n_components = 4)
X = pca.fit_transform(X)
X = pd.DataFrame(X)
```

```python
clf = SVC(C = 2**32, kernel = 'linear')
clf.fit(X, Y)
```

**NOTE: FULL SCORE MEANS LINEARLY SEPARABLE IN CURRENT FEATURES.**

```python
clf.score(X, Y)
```
